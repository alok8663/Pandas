# Write CSV File

```python
import pandas as pd
dis={'a':[1,2,3,4,5,6],'s':[6,5,4,3,2,1],'d':[12,3,4,6,5,7]}
d=pd.DataFrame(dis)
print(d)

d.to_csv("Test_new.csv",header=['A','B','C'],index=False)
```
