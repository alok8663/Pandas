# dropna and fillna

```python
import pandas as pd
var=pd.read_csv('Data.csv')
var
```

```python
var1=var.dropna() # This will drop all the rows containing NaN.
var1
```

```python
var2=var.dropna(axis=0) # This will drop all the rows containing NaN.
var2
```

```python
var3=var.dropna(axis=1) # This will drop all the columns containing NaN.
var3
```

```python
var4=var.dropna(subset=["Age"]) # This will drop that row containing NaN which is present in Age column.
var4
```

```python
var.dropna(inplace=True)
var
```

# fillna

```python
dataset=pd.read_csv('Data.csv')
dataset
```

```python
dataset1=dataset.fillna('python')
dataset1
```

```python
dataset2=dataset.fillna(method='ffill')
dataset2
```

```python
dataset3=dataset.fillna(method='bfill')
dataset3
```

```python
dataset.fillna(13,inplace=True)
dataset
```

# Replace

```python
ds=pd.read_csv('Data.csv')
ds
```

```python
ds.replace(to_replace=['No','Yes'],value=[0,1])
```

```python
ds.replace(to_replace='France',value='India')
```

```python
ds.replace(['Yes','No'],33)
```

```python
ds.replace("[A-Za-z]",'python',regex=True)
```
